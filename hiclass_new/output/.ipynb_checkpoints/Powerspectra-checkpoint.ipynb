```python
z = [0,0.1,0.2]
mu_lin7_class = [[] for j in range(len(z))]
mu2_lin = [[] for j in range(len(z))]
M_lin = Class()
k_pivot = 0.05

linear ={
        "Omega_scf": 0,
        "Omega_Lambda": 0,
        "cs2_fld": 1e-7,
        "w0_fld": -1,
#         "wa_fld":-0.5,
#         "c_gamma_over_c_fld": 0.1,
        "h": 0.67556,
        "omega_b": 0.022032,
        "omega_cdm": 0.12038,
        "A_s": 2.215e-9,
        "n_s": 0.9619,
        "z_max_pk": 10,
        "tau_reio": 0.05430842,
        "N_ur":3.046,
        "output": "mTk,mPk",
        "use_ppf":"yes",
        "P_k_max_1/Mpc": 130,
    }

M_lin.set(linear)

M_lin.compute()

h = M_lin.h()

As = linear["A_s"]
ns = linear["n_s"]
h = M_lin.h()


backgrounds = M_lin.get_background()

f_rho_b = interp1d(backgrounds['z'], backgrounds['(.)rho_b'], kind='cubic')
f_rho_cdm = interp1d(backgrounds['z'], backgrounds['(.)rho_cdm'], kind='cubic')
f_rho_fld = interp1d(backgrounds['z'], backgrounds['(.)rho_fld'], kind='cubic')
f_rho_tot = interp1d(backgrounds['z'], backgrounds['(.)rho_tot'], kind='cubic')



for zind, zi in enumerate(z):

    rho_b = f_rho_b(zi)
    rho_cdm = f_rho_cdm(zi)
    rho_fld = f_rho_fld(zi)
    rho_tot = f_rho_tot(zi)

    rho_m = rho_b + rho_cdm 

    Transfers = M_lin.get_transfer(zi)

    

    f_delta_fld = interp1d(Transfers['k (h/Mpc)'], Transfers['d_fld'], kind='cubic')
    f_delta_b = interp1d(Transfers['k (h/Mpc)'], Transfers['d_b'], kind='cubic')
    f_delta_cdm = interp1d(Transfers['k (h/Mpc)'], Transfers['d_cdm'], kind='cubic')
    f_delta_tot = interp1d(Transfers['k (h/Mpc)'], Transfers['d_tot'], kind='cubic')
    f_phis = interp1d(Transfers['k (h/Mpc)'], Transfers['phi'], kind='cubic')



    for k in kk:

        delta_fld = (f_delta_fld(k))
        delta_b =  f_delta_b(k)
        delta_cdm =  f_delta_cdm(k)
        delta_tot =  f_delta_tot(k)
        
        rhodDE_class = rho_tot*delta_tot - rho_cdm*delta_cdm - rho_b*delta_b
#         print(rhodDE_class)

        Pk_lin = (M_lin.pk(k*h,zi)*h**3)

        delta_m_lin =((Pk_lin/(2*np.pi**2))/(As*(k**(-3))*(k*h/k_pivot)**(ns-1)))**(1/2)
        
        phi = f_phis(k)
    
#         delta_m_lin = (rho_b*f_d_b[-1](k))/rho_m + (rho_cdm*f_d_cdm[-1](k))/rho_m
    
        mu2_lin[zind].append(((k*h)**4)*(phi**2)/((1.5*(a(zi)**2)*rho_m)**2*(delta_m_lin**2)))
        
        mu_lin7_class[zind].append(((1 - (delta_fld* rho_fld )/(rho_m*delta_m_lin))))
#         mu_lin7_class[zind].append(((1 - (rhodDE_class)/(rho_m*delta_m_lin))))
```

# Propto_omega

```python
mu_lin7_prop_omega = [[] for j in range(len(z))]
mu2_propto_omega_hi = [[] for j in range(len(z))]

M_lin = Class()


linear = {
    "Omega_smg": -1,
    "Omega_Lambda": 0,
    "Omega_fld": 0,
    "h": 0.67556,
    "omega_b": 0.022032,
    "omega_cdm": 0.12038,
    "A_s": 2.215e-9,
    "n_s": 0.9619,
    "z_max_pk": 10,
    "tau_reio": 0.05430842,
    "N_ur": 3.046,
    "gravity_model": "propto_omega",
    "parameters_smg": f"3.000000e+06, 0, 0, 0, 1",
    "expansion_model": "w0wa",
    "expansion_smg": f"6.878623e-01,-0.9,0",
    "output": "dTk,vTk,mPk",
    "P_k_max_1/Mpc": 130,
}

M_lin.set(linear)
M_lin.compute()
gmodel = linear["gravity_model"]

As = linear["A_s"]
ns = linear["n_s"]
h = M_lin.h()


backgrounds_hi = M_lin.get_background()


f_rho_b_hi = interp1d(backgrounds_hi["z"], backgrounds_hi["(.)rho_b"], kind="cubic")
f_rho_cdm_hi = interp1d(backgrounds_hi["z"], backgrounds_hi["(.)rho_cdm"], kind="cubic")
f_rho_smg_hi = interp1d(backgrounds_hi["z"], backgrounds_hi["(.)rho_smg"], kind="cubic")
f_rho_tot_hi = interp1d(backgrounds_hi["z"], backgrounds_hi["(.)rho_tot"], kind="cubic")
f_H_hi = interp1d(backgrounds_hi["z"], backgrounds_hi["H [1/Mpc]"], kind="cubic")
cs2_hi = interp1d(backgrounds_hi["z"], backgrounds_hi["c_s^2"], kind="cubic")


for zind, zi in enumerate(z):

    rho_b_hi = f_rho_b_hi(zi)
    rho_cdm_hi = f_rho_cdm_hi(zi)
    rho_smg_hi = f_rho_smg_hi(zi)
    rho_tot_hi = f_rho_tot_hi(zi)
    H_hi = f_H_hi(zi)

    rho_m_hi = rho_b_hi + rho_cdm_hi

    Transfers_hi = M_lin.get_transfer(zi)

    f_vx_smg = interp1d(Transfers_hi["k (h/Mpc)"], Transfers_hi["vx_smg"], kind="cubic")
    f_vx_prime_smg = interp1d(
        Transfers_hi["k (h/Mpc)"], Transfers_hi["vx_prime_smg"], kind="cubic"
    )
    f_delta_b_hi = interp1d(
        Transfers_hi["k (h/Mpc)"], Transfers_hi["d_b"], kind="cubic"
    )
    f_delta_cdm_hi = interp1d(
        Transfers_hi["k (h/Mpc)"], Transfers_hi["d_cdm"], kind="cubic"
    )
    f_delta_tot_hi = interp1d(
        Transfers_hi["k (h/Mpc)"], Transfers_hi["d_tot"], kind="cubic"
    )
    f_phis_hi = interp1d(Transfers_hi["k (h/Mpc)"], Transfers_hi["phi"], kind="cubic")

    for k in kk:

        delta_smg = -((1 - 0.9) / cs2_hi(zi)) * (
            -(3 * cs2_hi(zi) - 1) * a(zi) * H_hi * f_vx_smg(k) + f_vx_prime_smg(k))
        delta_b_hi = f_delta_b_hi(k)
        delta_cdm_hi = f_delta_cdm_hi(k)
        delta_tot_hi = f_delta_tot_hi(k)

        rhodDE = (
            rho_tot_hi * delta_tot_hi
            - rho_cdm_hi * delta_cdm_hi
            - rho_b_hi * delta_b_hi
        )
        #         print(delta_smg )

        Pk_lin = M_lin.pk(k * h, zi) * h**3

        delta_m_lin_hi = (
            (Pk_lin / (2 * np.pi**2))
            / (As * (k ** (-3)) * (k * h / k_pivot) ** (ns - 1))
        ) ** (1 / 2)

        phi = f_phis_hi(k)

        mu2_propto_omega_hi[zind].append(
            ((k * h) ** 4)
            * (phi**2)
            / ((1.5 * (a(zi) ** 2) * rho_m_hi) ** 2 * (delta_m_lin_hi**2))
        )

        mu_lin7_prop_omega[zind].append(
            ((1 - (delta_smg * rho_smg_hi) / (rho_m_hi * delta_m_lin_hi)))
        )

        # mu_lin7_prop_omega[zind].append(((1 - (rhodDE)/(rho_m_hi*delta_m_lin))))
```

```python


alphaB = [0,0.625,1.25,1.875,2.5]

Pk_alphaB = [[] for _ in range(len(alphaB))]
CMB_alphaB = [[] for _ in range(len(alphaB))]

for ind,val in enumerate(alphaB):
    
    M_lin = Class()

    linear =     {
        "Omega_smg": -1,
        "Omega_Lambda": 0,
        "Omega_fld": 0,
        "gravity_model": "propto_omega",
        "parameters_smg": f"1, {alphaB[ind]}, 0, 0, 1",
        "expansion_model" :"lcdm",
        "z_max_pk": 3,
        "output": "tCl,mTk,mPk,lCl",
        'lensing':'yes',
        "P_k_max_h/Mpc": 130,
    }

    M_lin.set(linear)

    M_lin.compute()
    
    h = M_lin.h()
    
    for k in kk:
        Pk_alphaB[ind].append(M_lin.pk(k*h,0)*h**3)
        
    cls = M_lin.lensed_cl(2500)

    ll = cls['ell'][2:]
    clTT = cls['tt'][2:]
    
    CMB_alphaB[ind].append(clTT*ll*(ll+1)/2./(np.pi))
```
